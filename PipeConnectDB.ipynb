```python
from sqlalchemy import create_engine
import pandas as pd
#Database Type: The example uses PostgreSQL. For other databases, you may need to install the appropriate connector (pyodbc for SQL Server, mysql-connector-python for MySQL, etc.).
```

```python
# Define your database connection details
db_user = 'your_username'
db_password = 'your_password'
db_host = 'localhost'  # Replace with your host, e.g., '127.0.0.1'
db_port = '5432'       # Replace with your port, e.g., '3306' for MySQL, '1433' for SQL Server
db_name = 'your_database'
```

```python
# Create the database connection URL
db_url = f'postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}'
```

```python
# Create the SQLAlchemy engine
engine = create_engine(db_url)
```

```python
# SQL query to select data from your table
query = "SELECT * FROM your_table_name"
```

```python
# Load data into a DataFrame
df = pd.read_sql(query, engine)
```
